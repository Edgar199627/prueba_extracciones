# Procedimiento 

```python

headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36"}
url = "https://es.wikipedia.org/wiki/F%C3%B3rmula_1"
response = requests.get(url, headers=headers)
# StringIO convierte el texto en un "archivo virtual" que read_html puede leer
tablas = pd.read_html(StringIO(response.text), flavor="html5lib")
#print(f"Total de tablas encontradas: {len(tablas)}")
pilotos = tablas[12]
```

```python
pilotos.to_csv(f'{camino_output}/pilotos.csv',index=False)
```
